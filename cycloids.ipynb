```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw_epicycloid(a, b, n=200, alpha=0.4, linestyle='g-'):
    '''
    Function to draw epicycloids. 
    See 
    https://en.wikipedia.org/wiki/Epicycloid
    https://mathworld.wolfram.com/Epicycloid.html
    Inspired by/based on 
    https://www.johndcook.com/blog/2020/04/24/envelopes-of-epicycloids/ 
    '''
    x = np.linspace(0, 2 * np.pi, n)
    plt.axes().set_aspect(1)
    plt.rcParams['axes.facecolor'] = 'black'
    plt.axis('off')
    plt.plot([np.cos(a*x), np.cos(b*x)], [np.sin(a*x), np.sin(b*x)], linestyle, alpha=alpha)
    plt.show()
```

```python
draw_epicycloid(1,10)
```

```python
def create_multi_cycloids(avals, bvals, n=200, alpha=0.4, color="green", backcolor=(0,0,0)):
    # Create a figure:
    plt.figure(figsize=(20, 20),facecolor=backcolor)
    for i, a in enumerate(avals):
        # for each subplot load in the ith frame
        ax = plt.subplot(int(np.sqrt(len(avals))), int(np.sqrt(len(avals))), i+1)
        x = np.linspace(0, 2 * np.pi, n)
        ax.set_aspect(1)
        plt.axis('off')
        ax.plot([np.cos(avals[i]*x), np.cos(bvals[i]*x)], [np.sin(avals[i]*x), np.sin(bvals[i]*x)], color, alpha=alpha)
    # Save the set of subplots:
    plt.savefig("cycloids.png")
```

```python
create_multi_cycloids([1,1,1,2,2,2,3,3,3], [2, 3, 4, 3, 5, 7, 4, 5, 7], color='xkcd:brownish pink', backcolor='xkcd:pale')
```

```python
def draw_rose(k, n=200, alpha=0.4, color='xkcd:deep rose'):
    '''
    Function to draw epicycloids. 
    See 
    https://en.wikipedia.org/wiki/Epicycloid
    https://mathworld.wolfram.com/Epicycloid.html
    Inspired by/based on 
    https://www.johndcook.com/blog/2020/04/24/envelopes-of-epicycloids/ 
    '''
    x = np.linspace(0,  20 * np.pi, n)
    plt.axes().set_aspect(1)
    plt.rcParams['axes.facecolor'] = 'black'
    plt.axis('off')
    plt.plot(np.cos(k*x)*np.cos(x), np.cos(k*x)*np.sin(x), color, alpha=alpha)
    plt.show()
```

```python
draw_rose(6/7)
```

```python
def create_multi_roses(kvals, n=200, alpha=0.4, color='xkcd:deep rose', backcolor='xkcd:pale'):
    # Create a figure:
    plt.figure(figsize=(20, 20),facecolor=backcolor)
    for i, a in enumerate(kvals):
        # for each subplot load in the ith frame
        ax = plt.subplot(3, 3, i+1)
        x = np.linspace(0, 30 * np.pi, n)
        ax.set_aspect(1)
        plt.axis('off')
        k = kvals[i]
        if i not in [1, 3,4,5, 7]:
            ax.plot(np.cos(k*x)*np.cos(x), np.cos(k*x)*np.sin(x), color, alpha=alpha)
    # Save the set of subplots:
    plt.savefig("roses.png")
```

```python
create_multi_roses([6/7,2,6/7,4,5,6,6/7,8,6/7], n=1000, backcolor='xkcd:pale')
```
